```python
df['Processor Name']=df['Processor'].apply(lambda x:' '.join(x.split()[0:3]))
```

```python
df
```

```python
def fetch_processor(text):
    if text == 'AMD Ryzen 5' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df
```

```python
df['Processor brand'] = df['Processor Name'].apply(fetch_processor)
```

```python
df['Processor brand'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df['Processor brand'],y=df['Price'])
plt.show()
```
