```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.flipkart.com/computers/laptops/pr?sid=6bo%2Cb5g&otracker=categorytree&page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
containers = soup.find_all('div',class_='_2kHMtA')
```

```python
len(containers)
```

```python
#data of first container
containers[0]
```

```python
Company1=[]
Rating1=[]
No_rating1=[]
Review1=[]
Price1=[]
MRP1=[]
Processor1=[]
RAM1=[]
OpSys1=[]
Memory1=[]
Size1=[]
ImageURL1=[]

for container in containers:
        
        company = container.findAll('div',{'class':'_4rR01T'})
        try:
            Company1.append(company[0].text.strip().split()[0])
        except:
            Company1.append(None)
        
        
        rating= container.find('div',{'class':'_3LWZlK'})
        try:
             Rating1.append(rating.text)
        except:
            Rating1.append(None)
        
        
        no_rating=container.find('span',{'class':'_2_R_DZ'})
        try:    
            ratRev=re.findall('\d+,?\d*,*',no_rating.text)
            no_rating=ratRev[0]
            No_rating1.append(no_rating)
        except:
             No_rating1.append(None)
                
        
        review=container.find('span',{'class':'_2_R_DZ'})
        try:    
            ratRev=re.findall('\d+,?\d*,*',review.text)
            review=ratRev[1]
            Review1.append(review)
        except:
             Review1.append(None)
                
        
        price=container.find('div',{'class':'_30jeq3 _1_WHN1'})
        try:
            Price1.append(price.text)
        except:
            Price1.append(None)

        mrp=container.find('div',{'class':'_3I9_wc _27UcVY'})
        try:
            MRP1.append(mrp.text)
        except:
            MRP1.append(None)
            
        
        info = container.findAll('li',{'class':'rgWa7D'})
        try:
            Processor1.append(info[0].text)
        except:
            Processor1.append(None)
            
            
        info = container.findAll('li',{'class':'rgWa7D'})
        try:
            match = re.search(r'\d+\sGB',info[1].text )
            RAM1.append(match.group())
        except:
            RAM1.append(None)
            
        
        info = container.findAll('li',{'class':'rgWa7D'})
        try:
            opsys=re.sub(r'\bOperating System\b', '', info[2].text)
            OpSys1.append(opsys)
        except:
            OpSys1.append(None)
            
            
        info = container.findAll('li',{'class':'rgWa7D'})
        try:
            Memory1.append(info[3].text)
        except:
            Memory1.append(None)
             
            
        info = container.findAll('li',{'class':'rgWa7D'})
        try:
            match = re.search(r'\d+.?\d*\scm',info[4].text )
            Size1.append(match.group())
        except:
            Size1.append(None)
                
        
        ImageURL = container.img
        try:
            ImageURL1.append(ImageURL.get('src'))
        except:
            ImageURL1.append(None)
            
            
d={'Company':Company1,'Rating':Rating1,'No_of_ratings':No_rating1,'Review':Review1,'Size':Size1,'Processor':Processor1,'RAM':RAM1,'Memory':Memory1,'OpSys':OpSys1,'Price':Price1,'MRP':MRP1,'ImgURL':ImageURL1}           
            
df=pd.DataFrame(d)
df.shape
```
